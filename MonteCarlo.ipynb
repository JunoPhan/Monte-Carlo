```python
#generating one possible future value based on market history using the expected return 9% with 18% volatiloty
import numpy as np
import locale
locale.setlocale(locale.LC_ALL, '')

pv = 10000
annual_add= 10000
vola =.18
ex_return = .09
time = 30
print("\t Return","\t Ending value".rjust(18))
for year in range(time):
    market_return = np.random.normal(ex_return,vola)
    fv = pv * (1+market_return) + annual_add 
    print("\t{}".ljust(10).format(round(market_return,4)),"\t{}".ljust(10).format(locale.currency(fv,grouping=True)))
    pv = fv
```

```python
#Simulate portfolio market value
import locale
locale.setlocale(locale.LC_ALL, '')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

sim = pd.DataFrame()

it = 200

bigtuple = [] #list of tuples
for x in range(it):
    expect_return = 0.09
    vola = 0.18
    time = 30
    pv = 10000
    add = 10000
    stream = []
    for i in range(time):
        rate = np.random.normal(expect_return, vola)
        end = round(pv * (1 + rate) + add,2)
        #stream.append(locale.currency(end, grouping = True))
        stream.append(end)
        pv = end
        smalltuples = (round(rate,4),end)  #tuple includes pairs of market interest rate and ending value
        #smalltuples.append(round(rate,4))
        #smalltuples.append(locale.currency(ending, grouping = True))
        bigtuple.append(smalltuples)
       
    sim[x] = stream
df = pd.DataFrame(bigtuple, columns = ["market interest rate","ending value"])
#print(stream)


print("tuple rate and ending value: ")
print(bigtuple)
print(len(bigtuple))
print(df.head())

#4 Summary statistics with numpy function

print(type(df["ending value"]))
df["ending value"].astype(float,False,'raise')   #convert string series to numeric type
print("Summary statistics with numpy functions")
print("COUNT: ", len(bigtuple))
print("MEAN : ",locale.currency(round(np.mean(df["ending value"]),2),grouping = True))
print("SD   : ",locale.currency(round(np.std(df["ending value"]),2),grouping = True))
print("MAX  : ", locale.currency(round(np.max(df["ending value"]),2),grouping = True))
print("MIN  : ", locale.currency(round(np.min(df["ending value"]),2),grouping = True))

print(sim[list(range(5))])

plt.plot(sim[list(range(5))])
plt.show(list(range(5)))
```
